```python
import os
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt

DATA_DIR = r"D:\CSE-Senior 2\GP\all data\After preprocessing\landmarks_preprocssed_with_SW_and_mask"

samples = []
labels = []
shapes = []

for file in os.listdir(DATA_DIR):
    if file.endswith(".npy") and not file.endswith("_mask.npy"):
        file_path = os.path.join(DATA_DIR, file)
        data = np.load(file_path)
        
        samples.append(file)
        shapes.append(data.shape)
        
        label = file.split("_")[0]
        labels.append(label)

total_samples = len(samples)
unique_classes = len(set(labels))
class_counter = Counter(labels)

print("===== DATASET ANALYSIS =====")
print("Total Samples:", total_samples)
print("Total Classes:", unique_classes)
print("Unique Data Shapes:", set(shapes))

print("\nTop 10 Largest Classes:")
for cls, count in class_counter.most_common(10):
    print(cls, ":", count)

print("\nTop 10 Smallest Classes:")
for cls, count in sorted(class_counter.items(), key=lambda x: x[1])[:10]:
    print(cls, ":", count)

plt.figure()
plt.hist(class_counter.values(), bins=30)
plt.title("Class Frequency Distribution")
plt.xlabel("Samples per Class")
plt.ylabel("Number of Classes")
plt.show()
```
