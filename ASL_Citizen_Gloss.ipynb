```python
import pandas as pd
from collections import Counter
import re

# Load CSV
df = pd.read_csv("/kaggle/input/asl-citizen/ASL_Citizen/splits/train.csv")

def normalize_gloss(gloss):
    """
    Normalize gloss by:
    - Lowercasing
    - Removing any trailing numbers (like ROSE2 -> rose)
    - Removing spaces
    """
    gloss = gloss.lower()
    gloss = re.sub(r'\d+$', '', gloss)  # remove trailing digits
    gloss = gloss.replace(" ", "")      # remove spaces
    return gloss

# Apply normalization
df['normalized_gloss'] = df['Gloss'].apply(normalize_gloss)

# Count occurrences
gloss_counts = Counter(df['normalized_gloss'])

# Convert to DataFrame
gloss_counts_df = pd.DataFrame(gloss_counts.items(), columns=['gloss', 'count'])

# Sort by count descending (optional)
gloss_counts_df = gloss_counts_df.sort_values(by='count', ascending=False)

# Save to CSV
gloss_counts_df.to_csv("asl_citizen_gloss_counts.csv", index=False)

print("Saved unique gloss counts to asl_citizen_gloss_counts.csv")
```
