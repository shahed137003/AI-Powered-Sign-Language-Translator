```python
import pandas as pd
import numpy as np
```

```python
df_train=pd.read_csv('/kaggle/input/asl-citizen/ASL_Citizen/splits/train.csv')
df_test=pd.read_csv('/kaggle/input/asl-citizen/ASL_Citizen/splits/test.csv')
df_val=pd.read_csv('/kaggle/input/asl-citizen/ASL_Citizen/splits/val.csv')
df_train.head()
```

```python
df_test.head()
```

```python
df_val.head()
```

```python
merged_df=pd.concat([df_train,df_test,df_val],axis='rows',ignore_index=True)
merged_df.head()
```

```python
merged_df.tail()
```

```python
import re #re is regular expressions
def cleandf(x):
    x=x.strip().lower()
    x=re.sub('\d+$','',x) #sub means substitute, \d means digit, + means 1 or more digit, $ means at the end
    #so it substitutes any digit at the end by nothing, ie removes it
    return x
print(cleandf("SoCCer2"))
```

```python
merged_df.head()
```

```python
merged_df['Gloss']=merged_df['Gloss'].apply(lambda x: cleandf(x))
merged_df.head()
```

```python
data_gloss_count = (
    merged_df.groupby("Gloss")
      .agg(
          count=("Gloss", "size"),                 
          videos=("Video file", lambda x: ','.join(x))   # join video filenames with comma
      )
      .reset_index()
      .sort_values(by="count", ascending=False)
)

print(data_gloss_count.head(50))
```
