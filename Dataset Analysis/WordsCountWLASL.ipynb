```python
import os
from collections import Counter

folder_path = r"E:\WLASL\landmarks_mapped\landmarks_mapped"
output_file = r"E:\WLASL\wlasl_word_counts.txt" 

word_counts = Counter()

for file in os.listdir(folder_path):
    if file.endswith(".npy"):
        # Remove extension
        name = os.path.splitext(file)[0].strip()
        # Stop at first underscore if exists
        base_name = name.split("_", 1)[0]
        word_counts[base_name] += 1
        
        # Sort by count descending
        
        sorted_words = word_counts.most_common()
    
    # Print and save to text file

with open(output_file, "w", encoding="utf-8") as f:
    for word, count in sorted_words:
        line = f"{word}: {count}"
        print(line)
        f.write(line + "\n")
```
