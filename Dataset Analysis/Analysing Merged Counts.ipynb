```python
from collections import Counter

file1 = r"E:\WLASL\wlasl_word_counts.txt"
file2 = r"E:\WLASL\msasl_words_count.txt" # output_msasl.txt
output_file = r"E:\WLASL\merged_counts.txt"

def read_counts(file_path):
    counts = Counter()
    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            if ": " in line:
                word, count = line.split(": ")
                counts[word] = int(count)
    return counts

# Read both files

counts1 = read_counts(file1)
counts2 = read_counts(file2)

# Merge counts

merged_counts = counts1 + counts2

# Sort descending

sorted_counts = merged_counts.most_common()

# Save merged counts

with open(output_file, "w", encoding="utf-8") as f:
    for word, count in sorted_counts:
        f.write(f"{word}: {count}\n")

# Compare words

words1 = set(counts1.keys())
words2 = set(counts2.keys())

common_words = words1 & words2
unique_to_file1 = words1 - words2
unique_to_file2 = words2 - words1

print(f"Total words in file1: {len(words1)}")
print(f"Total words in file2: {len(words2)}")
print(f"Common words: {len(common_words)}")
print(f"Unique to file1: {len(unique_to_file1)}")
print(f"Unique to file2: {len(unique_to_file2)}")


print("\nUnique words in file2 with their counts:")
for word in unique_to_file2:
    print(f"{word}: {counts2[word]}")
```
