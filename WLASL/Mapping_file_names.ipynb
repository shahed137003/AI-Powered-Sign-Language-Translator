```python
import os
import json
import numpy as np
```

```python
npy_folder = "landmarks2"          # folder with your original .npy files
mapped_folder = "landmarks_mapped" # folder to save mapped files
```

```python
json_path = "WLASL_v0.3.json"
with open(json_path, 'r') as f:
    data = json.load(f)
```

```python
video_to_word = {}
for gloss_entry in data:
    gloss = gloss_entry['gloss']
    for instance in gloss_entry['instances']:
        vid = str(instance['video_id'])
        video_to_word[vid] = gloss
```

```python
for npy_file in os.listdir(npy_folder):
    if not npy_file.endswith(".npy"):
        continue
    
    base_name = os.path.splitext(npy_file)[0]  # get video basename
    
    # Map to word
    word = video_to_word.get(base_name, None)
    if word is None:
        print(f"No mapping found for {base_name}, skipping...")
        continue
    
    # Load existing npy file
    arr = np.load(os.path.join(npy_folder, npy_file))
    
    # Prepare save path with word as filename
    save_name = f"{word}.npy"
    save_path = os.path.join(mapped_folder, save_name)
    
    # Handle duplicates
    counter = 1
    while os.path.exists(save_path):
        save_path = os.path.join(mapped_folder, f"{word}_{counter}.npy")
        counter += 1

    # Save the mapped file
    np.save(save_path, arr)
    print(f"Saved {save_path} with shape {arr.shape}")

print("All npy files mapped and saved successfully!")
```
